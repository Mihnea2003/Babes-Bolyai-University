***Tensor Shapes***

On performing operations on two or more tensors, they will need to be of the same shape - that is, having the same number of dimensions and the same number of cells in each dimension. For that, we have the `torch.*_like()` methods:


```python
x = torch.empty(2, 2, 3)
print(x.shape)
print(x)

empty_like_x = torch.empty_like(x)
print(empty_like_x.shape)
print(empty_like_x)

zeros_like_x = torch.zeros_like(x)
print(zeros_like_x.shape)
print(zeros_like_x)

ones_like_x = torch.ones_like(x)
print(ones_like_x.shape)
print(ones_like_x)

rand_like_x = torch.rand_like(x)
print(rand_like_x.shape)
print(rand_like_x)
```

## Task 6 -- An application: Simulate an unfair probability spinner

Conside an unfair probability spinner with $n$ slots. Each slot has a probabilty $P_n \in [0,1]$ and $\sum_{i=0}^{n-1}P_i = 1$. 

Write a function that random generate an index $i$ with the probability $P_i$ from the set $\{0, 1, ..., n\}$.  

```python
import numpy as np

def spinner(angleDistribution):
    # convert angle distribution to prob distribution
    total_angle = sum(angleDistribution)
    probabilityDistribution = np.array(angleDistribution) / total_angle
    
    random_number = np.random.random()
    
    # compute the cumulative sum of the prob distribution and find the index
    slot_index = np.searchsorted(np.cumsum(probabilityDistribution), random_number)
    
    return slot_index

angleDistribution = [72, 108, 36, 144]  # some angles
#print(spinner(angleDistribution))

def simulate_spinner(angleDistribution, num_simulations):
    counts = np.zeros(len(angleDistribution))
    for _ in range(num_simulations):
        index = spinner(angleDistribution)
        counts[index] += 1
    return counts

simulation_counts = simulate_spinner(angleDistribution, 1000)

for i, count in enumerate(simulation_counts):
    print(f"Angle {angleDistribution[i]}°: {count/10}% times")
```

## Task 7 -- Determine some outcomes using Monte Carlo Simulations

Consider a large pipe with a radius $R = 30 cm$ that has the opening pointing straight up, as in the figure. Around this pipe is a rectangular wall with an edge $l = 1m$. Balls of radius $r = 1 cm$ fall down uniformly from above inside the square yard. Compute, using a Monte Carlo simulation, the probability that a ball falls inside the pipe. Compare the results with the mathematically deduced probability.




```python
import numpy as np

R = 30  # radius of pipe
r = 1   # radius of balls 
l = 100  # side length of the square 
adjusted_R = R - r  # adjusted radius for the ball so it can enter the pipe

num_trials = 1_000_000
hits = 0

for _ in range(num_trials):
    x = np.random.uniform(0, l)
    y = np.random.uniform(0, l)
    
    distance_to_center = np.sqrt((x - l/2)**2 + (y - l/2)**2) # euclidean distance to the center of the square
    if distance_to_center <= adjusted_R:
        hits += 1

probability_simulation = hits / num_trials

probability_math = np.pi * adjusted_R**2 / l**2

probability_simulation, probability_math
```
